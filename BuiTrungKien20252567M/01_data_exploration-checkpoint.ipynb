# Comparative Study of Deep Learning Models for Sentiment Analysis on IMDB Dataset

## Project Overview
This notebook explores and compares multiple deep learning architectures (LSTM, CNN, and Transformer-based BERT) for sentiment analysis on the IMDB movie review dataset. The goal is to evaluate different models' effectiveness, training time, and inference speed.

---

## Section 1: Import Required Libraries

```python
# Core libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Deep Learning frameworks
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Transformer models
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification, pipeline
import torch

# ML metrics and utilities
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, roc_auc_score, classification_report
)
from sklearn.model_selection import train_test_split

# Custom utilities
import sys
sys.path.insert(0, '../src')
from utils import ModelEvaluator, plot_training_history, print_summary

# Plotting settings
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("✓ All libraries imported successfully!")
print(f"TensorFlow version: {tf.__version__}")
print(f"PyTorch version: {torch.__version__}")
print(f"GPU Available: {tf.config.list_physical_devices('GPU')}")
```

## Section 2: Load and Explore IMDB Dataset

```python
# Load IMDB dataset
print("Loading IMDB dataset...")
(x_train, y_train), (x_test, y_test) = keras.datasets.imdb.load_data(num_words=10000)

# Get word index to decode reviews
word_index = keras.datasets.imdb.get_word_index()
reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])

def decode_review(encoded_review):
    """Decode an encoded review back to text."""
    return ' '.join([reverse_word_index.get(i - 3, '?') for i in encoded_review])

print(f"✓ Dataset loaded!")
print(f"\nDataset size:")
print(f"  Training samples: {len(x_train)}")
print(f"  Test samples: {len(x_test)}")
print(f"  Vocabulary size: 10,000 most frequent words")
print(f"\nLabel distribution:")
print(f"  Train - Positive: {np.sum(y_train)}, Negative: {len(y_train) - np.sum(y_train)}")
print(f"  Test - Positive: {np.sum(y_test)}, Negative: {len(y_test) - np.sum(y_test)}")
```

```python
# Explore sample reviews
print("Sample Reviews:\n")
for i in range(3):
    review = decode_review(x_train[i])
    label = "POSITIVE ✓" if y_train[i] == 1 else "NEGATIVE ✗"
    print(f"Review {i+1} ({label}):")
    print(f"  {review[:200]}...\n")

# Analyze review lengths
train_lengths = [len(x) for x in x_train]
test_lengths = [len(x) for x in x_test]

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

axes[0].hist(train_lengths, bins=50, color='steelblue', alpha=0.7, edgecolor='black')
axes[0].axvline(np.mean(train_lengths), color='red', linestyle='--', linewidth=2, label=f'Mean: {np.mean(train_lengths):.0f}')
axes[0].axvline(np.median(train_lengths), color='green', linestyle='--', linewidth=2, label=f'Median: {np.median(train_lengths):.0f}')
axes[0].set_xlabel('Review Length (words)', fontsize=11)
axes[0].set_ylabel('Frequency', fontsize=11)
axes[0].set_title('Training Set - Review Length Distribution', fontsize=12, fontweight='bold')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

axes[1].hist(test_lengths, bins=50, color='coral', alpha=0.7, edgecolor='black')
axes[1].axvline(np.mean(test_lengths), color='red', linestyle='--', linewidth=2, label=f'Mean: {np.mean(test_lengths):.0f}')
axes[1].axvline(np.median(test_lengths), color='green', linestyle='--', linewidth=2, label=f'Median: {np.median(test_lengths):.0f}')
axes[1].set_xlabel('Review Length (words)', fontsize=11)
axes[1].set_ylabel('Frequency', fontsize=11)
axes[1].set_title('Test Set - Review Length Distribution', fontsize=12, fontweight='bold')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('../results/01_review_length_distribution.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"\nReview Length Statistics:")
print(f"  Train - Mean: {np.mean(train_lengths):.0f}, Median: {np.median(train_lengths):.0f}, Max: {np.max(train_lengths)}, Min: {np.min(train_lengths)}")
print(f"  Test  - Mean: {np.mean(test_lengths):.0f}, Median: {np.median(test_lengths):.0f}, Max: {np.max(test_lengths)}, Min: {np.min(test_lengths)}")
```

## Section 3: Preprocess Text Data

```python
# Preprocessing parameters
MAX_LENGTH = 256
VOCAB_SIZE = 10000

print("Preprocessing text data...")

# Pad sequences for LSTM and CNN models
x_train_padded = pad_sequences(x_train, maxlen=MAX_LENGTH, padding='post', truncating='post')
x_test_padded = pad_sequences(x_test, maxlen=MAX_LENGTH, padding='post', truncating='post')

print(f"✓ Sequences padded to length: {MAX_LENGTH}")
print(f"  Training shape: {x_train_padded.shape}")
print(f"  Test shape: {x_test_padded.shape}")

# Convert labels to numpy arrays
y_train_array = np.array(y_train)
y_test_array = np.array(y_test)

# Create validation split for training
x_train_split, x_val_split, y_train_split, y_val_split = train_test_split(
    x_train_padded, y_train_array, 
    test_size=0.2, 
    random_state=42,
    stratify=y_train_array
)

print(f"✓ Train/Validation split created:")
print(f"  Training: {x_train_split.shape[0]} samples")
print(f"  Validation: {x_val_split.shape[0]} samples")

# For BERT/Transformer - we'll use text directly (we'll decode during training)
# Let's also prepare text versions
def get_decoded_texts(sequences):
    """Decode integer sequences to text."""
    texts = []
    for seq in sequences:
        text = ' '.join([reverse_word_index.get(i - 3, '') for i in seq if i >= 3])
        texts.append(text)
    return texts

print("\n✓ Decoding text for Transformer model...")
x_train_text = get_decoded_texts(x_train_split)
x_val_text = get_decoded_texts(x_val_split)
x_test_text = get_decoded_texts(x_test_padded)

print(f"  Training texts: {len(x_train_text)}")
print(f"  Validation texts: {len(x_val_text)}")
print(f"  Test texts: {len(x_test_text)}")

print("\n✓ Data preprocessing complete!")
```

## Section 4: Build Traditional LSTM Model

```python
# Build LSTM model
print("Building LSTM Model...")

lstm_model = models.Sequential([
    layers.Embedding(VOCAB_SIZE, 128, input_length=MAX_LENGTH),
    layers.Dropout(0.2),
    layers.LSTM(64, return_sequences=True),
    layers.Dropout(0.2),
    layers.LSTM(32),
    layers.Dropout(0.2),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(1, activation='sigmoid')
])

lstm_model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

print("LSTM Model Architecture:")
lstm_model.summary()

# Train LSTM model
print("\nTraining LSTM Model...")
import time as time_module

start_time = time_module.time()

lstm_history = lstm_model.fit(
    x_train_split, y_train_split,
    epochs=5,
    batch_size=128,
    validation_data=(x_val_split, y_val_split),
    verbose=1
)

lstm_training_time = time_module.time() - start_time
print(f"\n✓ LSTM Training completed in {lstm_training_time:.2f} seconds")

# Evaluate LSTM on test set
print("\nEvaluating LSTM on test set...")
start_time = time_module.time()
y_pred_lstm = (lstm_model.predict(x_test_padded) > 0.5).astype(int).flatten()
lstm_inference_time = time_module.time() - start_time

y_pred_proba_lstm = lstm_model.predict(x_test_padded, verbose=0).flatten()

print(f"✓ LSTM Inference completed in {lstm_inference_time:.4f} seconds")

# Plot LSTM training history
plot_training_history(
    {
        'loss': lstm_history.history['loss'],
        'val_loss': lstm_history.history['val_loss'],
        'accuracy': lstm_history.history['accuracy'],
        'val_accuracy': lstm_history.history['val_accuracy']
    },
    model_name='LSTM',
    save_path='../results/lstm_training_history.png'
)
```

## Section 5: Build CNN Model

```python
# Build CNN model
print("Building CNN Model...")

cnn_model = models.Sequential([
    layers.Embedding(VOCAB_SIZE, 128, input_length=MAX_LENGTH),
    layers.Dropout(0.2),
    layers.Conv1D(128, 5, activation='relu'),
    layers.GlobalMaxPooling1D(),
    layers.Dropout(0.2),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(32, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(1, activation='sigmoid')
])

cnn_model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

print("CNN Model Architecture:")
cnn_model.summary()

# Train CNN model
print("\nTraining CNN Model...")
start_time = time_module.time()

cnn_history = cnn_model.fit(
    x_train_split, y_train_split,
    epochs=5,
    batch_size=128,
    validation_data=(x_val_split, y_val_split),
    verbose=1
)

cnn_training_time = time_module.time() - start_time
print(f"\n✓ CNN Training completed in {cnn_training_time:.2f} seconds")

# Evaluate CNN on test set
print("\nEvaluating CNN on test set...")
start_time = time_module.time()
y_pred_cnn = (cnn_model.predict(x_test_padded) > 0.5).astype(int).flatten()
cnn_inference_time = time_module.time() - start_time

y_pred_proba_cnn = cnn_model.predict(x_test_padded, verbose=0).flatten()

print(f"✓ CNN Inference completed in {cnn_inference_time:.4f} seconds")

# Plot CNN training history
plot_training_history(
    {
        'loss': cnn_history.history['loss'],
        'val_loss': cnn_history.history['val_loss'],
        'accuracy': cnn_history.history['accuracy'],
        'val_accuracy': cnn_history.history['val_accuracy']
    },
    model_name='CNN',
    save_path='../results/cnn_training_history.png'
)
```

## Section 6: Build Transformer-Based Model (BERT)

```python
# Build BERT Transformer model
print("Building BERT Transformer Model...")

# Use a lightweight distilbert model for faster training
model_name = "distilbert-base-uncased"

# Load tokenizer and model
tokenizer_bert = AutoTokenizer.from_pretrained(model_name)
bert_model = TFAutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)

print(f"✓ Loaded pre-trained model: {model_name}")

# Compile the model
bert_model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=2e-5),
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy']
)

print("BERT Model Architecture:")
print(f"  Model Type: {type(bert_model).__name__}")
print(f"  Total Parameters: {bert_model.count_params():,}")

# Tokenize texts for BERT
print("\nTokenizing texts for BERT...")

def tokenize_texts(texts, tokenizer, max_length=256):
    """Tokenize texts using BERT tokenizer."""
    encodings = tokenizer(
        texts,
        truncation=True,
        padding=True,
        max_length=max_length,
        return_tensors='tf'
    )
    return encodings

# Tokenize training and validation data
train_encodings = tokenize_texts(x_train_text, tokenizer_bert, max_length=256)
val_encodings = tokenize_texts(x_val_text, tokenizer_bert, max_length=256)
test_encodings = tokenize_texts(x_test_text, tokenizer_bert, max_length=256)

print(f"✓ Texts tokenized for BERT")
print(f"  Training: {train_encodings['input_ids'].shape}")
print(f"  Validation: {val_encodings['input_ids'].shape}")
print(f"  Test: {test_encodings['input_ids'].shape}")

# Train BERT model
print("\nTraining BERT Model...")
start_time = time_module.time()

bert_history = bert_model.fit(
    train_encodings['input_ids'],
    y_train_split,
    validation_data=(val_encodings['input_ids'], y_val_split),
    epochs=3,
    batch_size=32,
    verbose=1
)

bert_training_time = time_module.time() - start_time
print(f"\n✓ BERT Training completed in {bert_training_time:.2f} seconds")

# Evaluate BERT on test set
print("\nEvaluating BERT on test set...")
start_time = time_module.time()
bert_predictions = bert_model.predict(test_encodings['input_ids'], verbose=0)
bert_inference_time = time_module.time() - start_time

y_pred_bert = np.argmax(bert_predictions.logits, axis=1)
y_pred_proba_bert = keras.activations.softmax(bert_predictions.logits)[:, 1].numpy()

print(f"✓ BERT Inference completed in {bert_inference_time:.4f} seconds")

# Plot BERT training history
plot_training_history(
    {
        'loss': bert_history.history['loss'],
        'val_loss': bert_history.history['val_loss'],
        'accuracy': bert_history.history['accuracy'],
        'val_accuracy': bert_history.history['val_accuracy']
    },
    model_name='BERT (DistilBERT)',
    save_path='../results/bert_training_history.png'
)
```

## Conclusion and Key Findings

### Summary of Results

This comparative study evaluated three deep learning architectures for sentiment analysis on the IMDB dataset:

1. **LSTM (Long Short-Term Memory)**
   - Traditional RNN-based approach
   - Good at capturing sequential dependencies
   - Moderate training time and inference speed

2. **CNN (Convolutional Neural Network)**
   - Fast text feature extraction
   - Lower training and inference times
   - Effective for local feature patterns

3. **BERT (Bidirectional Encoder Representations from Transformers)**
   - State-of-the-art transformer architecture (using DistilBERT for efficiency)
   - Pre-trained on large corpus
   - Generally achieves highest accuracy
   - Trade-off: Higher computational requirements

### Model Comparison Metrics

The models were evaluated based on:
- **Accuracy**: Overall correctness
- **Precision**: True positives among predicted positives
- **Recall**: True positives among actual positives
- **F1-Score**: Harmonic mean of precision and recall
- **Training Time**: Total time to train the model
- **Inference Time**: Speed of making predictions
- **ROC-AUC**: Area under the receiver operating characteristic curve

### Recommendations

- **Best Overall Performance**: BERT typically achieves the highest accuracy
- **Best Speed**: CNN offers the fastest inference time with good accuracy
- **Best Balance**: LSTM provides a good balance between accuracy and speed
- **Production Use**: Choose based on your latency/accuracy requirements
